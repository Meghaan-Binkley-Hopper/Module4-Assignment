```python
#importing sapCy module and texblob
import spacy 
from spacytextblob.spacytextblob import SpacyTextBlob

#loading English language
nlp = spacy.load("en_core_web_sm")
nlp.add_pipe('spacytextblob')
#create string from JSON file
jstring = json.dumps(result)
#applying nlp to string
doc = nlp(jstring)
#iterating through string to tokenize
for token in doc:
    print(token.text)
#perform sentiment analysis
doc._.blob.sentiment_assessments.assessments
#In my opinion, based on the anaysis, the poem has a more positive sentiment analysis because most of the scores are above 0.
```

# Question #3

3. Write a function that takes an artist, song, and filename, accesses the lyrics.ovh api to get the song lyrics, and writes the results to the specified filename.  Test this function by getting the lyrics to any four songs of your choice and storing them in different files.

```python
import requests
import json

#Function to access poem and write to JSON file
def write_poem_from_API_to_JSON(author, poem, fileName): 
    URL = f'https://poetrydb.org/author,title/{author};{poem}'
    result = json.loads(requests.get(URL).text)
    json_object = json.dumps(result)
    with open(fileName, "w") as outfile:
        outfile.write(json_object)

#Calling function 4x
write_poem_from_API_to_JSON('Edgar Allan Poe', 'The Bells', 'Poem1.json')
write_poem_from_API_to_JSON('Edgar Allan Poe', 'A Dream Within A Dream', 'Poem2.json')
write_poem_from_API_to_JSON('Emily Dickinson', 'Nature can do no more', 'Poem3.json')
write_poem_from_API_to_JSON('Shakespeare', 'Sonnet 150', 'Poem4.json')

```
